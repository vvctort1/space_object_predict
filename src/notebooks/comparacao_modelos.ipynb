# Auto Machine Learning

Utilizando técnica de autoML com pycaret para encontrar melhores modelos de acordo com os dados de entrada.

```python
import pandas as pd
from pycaret.classification import *

df = pd.read_csv('../data/objetos_espaciais_complexo.csv')

exp = setup(
    data=df,
    target='tipo_objeto',
    session_id=42,
    train_size=0.7,
    normalize=True,
    fold=5,
    verbose=True,
    use_gpu=False
)

best_model = compare_models()

leaderboard = pull()
print(leaderboard.head(10))

tuned_model = tune_model(best_model)

final_model = finalize_model(tuned_model)

predictions = predict_model(final_model)
print(predictions.head())

# save_model(final_model, '../models/pycaret_space_objects_model')
```
